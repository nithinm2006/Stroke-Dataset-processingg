Question 6

```python
import matplotlib.pyplot as plt

# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Outliers Detected Using IQR:\n")

# Detect outliers using IQR
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} outliers")

# -------------------------------
# Box Plot for ID
# -------------------------------
plt.figure(figsize=(5,4))
plt.boxplot(df['id'])
plt.title("Box Plot of ID")
plt.ylabel("ID")
plt.show()

# -------------------------------
# Box Plot for Remaining Features
# -------------------------------
cols_without_id = numerical_cols.drop('id')

plt.figure(figsize=(10,5))
df[cols_without_id].boxplot()
plt.title("Box Plot of Remaining Numerical Features")
plt.xticks(rotation=45)
plt.ylabel("Values")
plt.show()
```

Question 7

```python

# Numerical columns (excluding 'id' and target 'stroke')
cols = ['age','avg_glucose_level', 'bmi']

# Handle outliers
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)

print("Outliers handled successfully.")

plt.figure(figsize=(10,5))
df[cols].boxplot()
plt.title("Box Plot After Handling Outliers")
plt.xticks(rotation=45)
plt.show()
```

Question 9

Min-Max Normalization

```python
from sklearn.preprocessing import MinMaxScaler

# Numerical columns
cols = ['age', 'avg_glucose_level', 'bmi']

# Create scaler
minmax = MinMaxScaler()

# Apply Min-Max Scaling
df_minmax = df.copy()
df_minmax[cols] = minmax.fit_transform(df_minmax[cols])

print("After Min-Max Normalization:")
display(df_minmax.head())
```

Z-score Scaling

```python
from sklearn.preprocessing import StandardScaler

# Create scaler
standard = StandardScaler()

# Apply Standardization
df_standard = df.copy()
df_standard[cols] = standard.fit_transform(df_standard[cols])

print("After Standardization:")
display(df_standard.head())
```
